```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame( # 시퀀스나 딕셔너리 형태
    np.arange(10, 22).reshape(3, 4), # 3행의 4열, 2차원 배열
    columns = ['A', 'B', 'C', 'D'], # 4열인데 3열에 대한 이름을 부여하면 에러 발생
    index = ['a','b','c'] # 3행인데 4행에 대한 이름을 부여하면 에러 발생
)
df
```

```python
#@title . loc 인덱서
#@markdown * 문자열 인덱스 -> 인덱싱, 슬라이싱
#@markdown * .loc[행 인덱스, 열 인데스], loc[행 인덱스]
#@markdown * 행 하나만 조회시 시리즈로 나온다.
# https://colab.research.google.com/notebooks/forms.ipynb
# row_name = 'a' #@param{"type":"string"}

row_name = 'a' #@param ["a", "b", "c"] {"type":"string"}
df.loc[row_name]
```

```python
# 슬라이싱 : 문자열 -> 끝점이 포함된다.
# df.loc['a':'b']
df.loc['c'::-1] # 증가폭도 가능
```

```python
#@title apply
df3 = pd.DataFrame(np.random.randint(1, 6, size=(5, 3)), columns=list('ABC'))
df3
```

```python
# 열의 최댓값 - 열의 최솟값
df3.max() - df3.min()
# df3.max(axis = 0) - df3.min(axis = 0)
```

```python
# df3.apply(lambda x: x.max() - x.min())
df3.apply(lambda x: x.max() - x.min(), axis = 0) 
# 행들에 대한 차원 반복,축소 연산 => 열을 기준으로 해서 나온다.
```

```python
df3.apply(lambda x: x.mean() - x.min(), axis = 1)
```

```python
df3.apply(lambda x: round(x.mean() - x.min()), axis = 1) # round : 반올림
```

```python
#@title 합성

customers = ['LEE', 'KIM', 'PARK', 'SON', 'SEO']
ids = [f"1{i:03d}" for i in range(len(customers))]
ids, customers
```

```python
df1 = pd.DataFrame({
    "id" : ids,
    "name" : customers,
})
df1
```

```python
df2 = pd.DataFrame({
    "id" : np.random.choice([f"1{i+1:03d}" for i in range(len(customers)+2)], 10, replace=True)
})
df2
```

```python
df2['money'] = (df2.id.apply(lambda x: int(x) * abs(np.random.randn()) * 1000) ).apply(int)
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='inner')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.merge(df1, df2, on='id')
```
